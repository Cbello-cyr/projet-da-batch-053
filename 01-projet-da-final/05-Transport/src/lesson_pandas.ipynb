```python
#merge == Jointure


# Let's create two new dataframes
meal = ['pizza', 'pasta', 'burger']
prices = [11.8, 12.9, 15.60]
calories = [870, 790, 950]

prices = pd.DataFrame({'meal': meal, 'price': prices})
calories = pd.DataFrame({'meal': meal, 'calories': calories})
```

```python
prices
```

```python
calories
```

```python
prices.merge(calories)
```

```python
prices.merge(calories, how='inner', left_on='meal', right_on='meal')
```

```python

# Let's create two new dataframes
meal = ['pizza', 'pasta', 'burger']
prices = [11.8, 12.9, 15.60]
calories = [870, 790, 950]

prices = pd.DataFrame({'meal': meal, 'price': prices})
calories = pd.DataFrame({'food': meal, 'calories': calories})
```

```python
prices
```

```python
calories
```

```python
joined = prices.merge(calories, how='inner', left_on='meal', right_on='food')
joined
```

# Save the DataFrame

```python
joined.to_csv('joined.csv')
```
